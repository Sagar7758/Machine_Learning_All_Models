### Linear Regression 
#### Problem Statement :
##### Predict a student's final exam score based on the number of hours they study 

```python
# importing the libraties and machine learning models (Linear Regression)
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Step No : 2 
# Creating the dataset 
sagardata = {"hours_study": [2,3,4,5,6,7,8,9,10],
             "Exam_score": [50,60,70,80,90,52,95,85,75]}

# Step NO : 3
# Creating DataFrame
df = pd.DataFrame(sagardata)
# print(df.to_string())

# Step No : 4 
# Splitting Input and Output 
x = df[["hours_study"]]
y = df[["Exam_score"]]

# Step No : 5 
# Scatter Plot 
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("Hours Study")
plt.ylabel("Exam Score")
plt.title("Hours Study vs Exam Score")
plt.show()

# Step NO : 6
# Train - test split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# Step No : 7
# Creating the Model 
model = LinearRegression()

# Step No : 8
# Training the  model 
model.fit(x_train , y_train)

# Step No : 9
# Regression line Visualization (Important for understanding the relationship between hours of study and exam score)

# Use the trained model to predict outputs for all input values
y_pred = model.predict(x)

import matplotlib.pyplot as plt
# Plot actual data points(blue dots )
# these represent real data (hours studied vs exam score )
plt.scatter(x,y,color='blue', label='Actual Data')

# Plot the regression line (red line)
plt.plot(x, y_pred, color='red', label='Regression Line')

# Label for X-axis (input feature )
plt.xlabel("Hours Study")
# Label for Y-axis (target variable )
plt.ylabel("Exam Score")

# Title of the graph for better understanding 
plt.title("Regression Line Visualization")

# Display the final graph
plt.show()

# Step No : 10
# Actual vs predicted visualization (Important for evaluating the model's performance and understanding how well it fits the data)
# Scatter Plot of Actual vs Predicted values
# Each point represtents how close the prediction is to the real value 
plt.scatter(y, y_pred)

# x-axis label (actual values )
plt.xlabel("Actual Exam Score")
# y-axis represents predicted values by the model 
plt.ylabel("Predicted Exam Score")
# Title of the graph for clarity
plt.title("Actual vs Predicted Exam Scores")
# Display the graph to evaluate the model's performance
plt.show()

# Step No : 11
user_input = float(input("Enter the number of hours you study : "))

# Step No 12
# Making Prediction 
predicted_score = model.predict([[user_input]])


# Step No : 13 
# Printing the output 
print(f"Predicted Exam Score : {predicted_score[0][0]:.2f}")
```

## Conclusion

- Linear Regression model successfully captured the relationship between hours studied and exam score
- The regression line shows a positive linear trend
- Actual vs predicted values indicate good model performance
- Some variation exists due to noise in the dataset
- Model can be improved using more data or feature engineering
