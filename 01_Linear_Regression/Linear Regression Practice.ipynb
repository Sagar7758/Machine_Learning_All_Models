### Linear Regression 
#### Problem Statement :
##### Predict a student's final exam score based on the number of hours they study 

```python
# importing the libraties and machine learning models (Linear Regression)
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Step No : 2 
# Creating the dataset 
sagardata = {"hours_study": [2,3,4,5,6,7,8,9,10],
             "Exam_score": [50,60,70,80,90,52,95,85,75]}

# Step NO : 3
# Creating DataFrame
df = pd.DataFrame(sagardata)
# print(df.to_string())

# Step No : 4 
# Splitting Input and Output 
x = df[["hours_study"]]
y = df[["Exam_score"]]

# Step NO : 5 
# Train - test split 
x_train , y_test , y_train , x_test = train_test_split(x,y, test_size=0.2 , random_state= 42)

# Step No : 6 
# Creating the Model 
model = LinearRegression()

# Step No : 7
# Training the  model 
model.fit(x_train , y_train)

# Step No : 8 
user_input = float(input("Enter the number of hours you study : "))

# Step No 9 : 
# Making Prediction 
predicted_score = model.predict([[user_input]])


# Step No : 10 
# Printing the output 
print(f"Predicted Exam Score : {predicted_score[0][0]:.2f}")
```
