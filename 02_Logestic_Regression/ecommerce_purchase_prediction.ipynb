### **Problem Statement :** 
#### In an e-commerce company, the management wants to predict whether a customer will purchase a high-value product based on their age, time spent on the website, and whether they have added items to their cart. The goal is to optimize marketing strategies by targeting potential customers more effectively, thereby increasing sales and revenue.

```python
# Step No : 1
# importing the libraries and machine learning models (Logistic Regression)
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Step No : 2
# Loading the dataset
df = pd.read_csv("ecommerce_data.csv")

# Step No : 3
# Understanding the Data 

# Displaying the first 5 rows of the dataset to understand structure and values
print("First 5 Rows:\n", df.head())

# Displaying the dataset information such as column names , data types and non-null counts to understand the structure of the dataset
print("\nDataset Information:")
df.info()

# Displaying the statistical summary of the dataset to understand the distribution of numerical features
print("\nStatistical Summary:")
df.describe()

# Checking duplicate rows in dataset
print("\nDuplicate Rows:", df.duplicated().sum())

# Step No : 4
# Checking for missing values in the dataset to identify any data quality issues
print("\nMissing Values:\n", df.isnull().sum())

# Step No : 5
# Spliting Inptut and Output features
x = df[["Age" , "Time_Spent" , "Added_to_Cart"]]
y = df[["Purchase"]]

# Step No : 6
# Visualizing the relationship between input features and target variable using scatter plots to identify any patterns or correlations
import matplotlib.pyplot as plt
plt.scatter(df["Time_Spent"], df["Purchase"], color='blue')
plt.title("Time Spent vs Purchased")
plt.xlabel("Time Spent (minutes)")
plt.ylabel("Purchased ")
plt.show()

# Step No : 7
# Splitting the dataset into training and testing sets to evaluate the performance of the machine learning model
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)

# Step No : 8
# Training the Logistic Regression model on the training data to learn the relationship between input features and target
model = LogisticRegression()

# Step No : 9
# Fitting the model to the training data to estimate the parameters of the logistic regression model
model.fit(x_train, y_train.values.ravel())

# Step No : 10
# Making predictions on the test set to evaluate the model's performance
y_pred = model.predict(x_test)

# Step No : 11
# Visualizing  (Actual vs Predicted) values using a scatter plot to assess the model's performance
plt.scatter(y_test,y_pred, color='red')
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.show()

# Step No : 12
# User Input for prediction
age = float(input("Enter Age:"))
time_spent = float(input("Enter Time Spent  : "))
cart = float(input("Added To Cart (0 or 1): "))

# Step No : 13
# Making a prediction based on user input to provide insights or recommendations
predicted = model.predict([[age, time_spent, cart]])

# Step No : 14
# Printing  Output
print("This is real output of the model prediction:", predicted[0])
if predicted[0] == 1:
    print("The user is likely to make a purchase.")
else:
    print("The user is unlikely to make a purchase.")
```

### ***Final Conclusion :***

In this project, a Logistic Regression model was developed to predict whether a user is likely to make a purchase based on features such as Age, Time Spent on the website, and Added to Cart status.

The dataset was explored and preprocessed to ensure data quality, including checking for missing values and duplicates. After splitting the dataset into training and testing sets, the model was trained successfully.

The model was able to learn the relationship between input features and the target variable and provide predictions based on new user input. The results demonstrate that Logistic Regression is effective for binary classification problems like purchase prediction.

Overall, the project successfully achieved its objective of building a predictive model and understanding the impact of user behavior on purchasing decisions.
